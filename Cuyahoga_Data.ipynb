## Finding a Good Location for a Gourmet Pizza Place in Cuyahoga County

### Data Sets

For this project I have located population and median income data for Cuyahoga County, OH. This data will be used to see if the demographics are suitable for a gourmet pizza place. This will be combined with Foursquare data to analyze the venues and see what cities have the least competition along with the right demographics. The data is shown below.

```python
import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

I obtained a download of latitude and longitude for all US cities at the following location:
https://simplemaps.com/data/us-cities. I removed the non-Ohio locations and uploaded the CSV. See header below.

```python
df1 = pd.read_csv('OHcities.csv')
df1.head()
```

I next obtained Ohio median income data from the following website and converted it to a CSV and uploaded it.  https://www.cleveland.com/datacentral/2016/12/median_family_income_household.html. See header below.

```python
df2 = pd.read_csv('OHcities_Income.csv')
df2.head()
```

Next, merge the dataframes and remove any city where the median income is $60,000 or less.

```python
df=df1.merge(df2)
df['median_income'] = df['median_income'].str.replace('$', '')
df['median_income'] = df['median_income'].str.replace(',', '')
df['median_income'] = df['median_income'].astype(float)
df=df.loc[df['median_income'] > 60000]
df
```

Finally, filter dataframe to include only cities in Cuyahoga County.

```python
dfcuyahoga=df[df['county'].str.contains("Cuyahoga")]
dfcuyahoga
```

This is the population we will analyze to find the best 2 cities for our 2 new pizza joints.
