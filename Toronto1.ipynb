# Toronto

```python
import pandas as pd
import requests
```

```python
wikipedia_link='https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
```

```python
toronto_wikipedia_page = requests.get(wikipedia_link)
```

```python
page=toronto_wikipedia_page.text
```

```python
df = pd.read_html(page)[0]
df
```

```python
df = df[df.Borough != 'Not assigned']
df
```

```python
df.Neighbourhood.replace(['Not assigned'], [df.Borough], inplace=True)
```

```python
df=df.sort_values(by=['Postcode'])
df=df.reset_index()
df
```

```python
dfnew=df
dfnew['Delete'] = 0
prevPostcode=''
prevNeighbourhood=''
for ind in dfnew.index: 
    #print(dfnew['Neighbourhood'][ind])   
    if prevPostcode==dfnew['Postcode'][ind]:
            neigh=dfnew['Neighbourhood'][ind]
            dfnew.iloc[ind, dfnew.columns.get_loc('Neighbourhood')] = neigh + ", " + prevNeighbourhood
            dfnew.iloc[ind-1, dfnew.columns.get_loc('Delete')] = 1
            #print(dfnew['Neighbourhood'][ind])
            #print(neigh + ", " + prevNeighbourhood)
    prevPostcode=dfnew['Postcode'][ind]
    prevNeighbourhood=dfnew['Neighbourhood'][ind]
```

```python
dfnew
```

```python
dfnew = dfnew[df.Delete != 1]
dfnew
```

```python
dfinal=dfnew.reset_index()
dfinal.drop(columns=['level_0', 'index', 'Delete'], inplace=True)
dfinal
```
